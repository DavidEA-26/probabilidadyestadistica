```python
import pandas as pd
import numpy as np
from statistics import mean, median, mode

maximo = autos2['price'].max()
minimo = autos2['price'].min()
rango = maximo - minimo
media = mean(autos2['price'])
mediana = median(autos2['price'])
moda = mode(autos2['price'])

print("Estadistica descrptiva de los autos", "Media =", media,
      "\n Mediana =", mediana, "\n Moda =", moda, "\n Maximo =", maximo,
      "\n Minimo =", minimo, "\n Rango =", rango)
```

```python
plt.pyplot.hist(autos2.price, bins = 7)

plt.pyplot.title("Histograma de precios de los automoviles")
plt.pyplot.xlabel("price")
plt.pyplot.ylabel("Frecuencia")
```
